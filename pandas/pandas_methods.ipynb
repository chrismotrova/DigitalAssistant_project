```python
import sqlite3
import pandas as pd
```

```python
file = 'data.xlsx'
df = pd.read_excel(file, sheet_name='Комиссия №1')
df.columns = [
    "3",
    "number",
    "id",
    "name",
    "group",
    "ze_rup",
    "ze",
    "1",
    "2",
    "mark_mentor",
    "mark_comission",
    "mark_total",
    "neyavka"
]
df = df.drop(labels=["1", "2", "3", "id", "number", "ze_rup", "ze", "neyavka"], axis=1)
df
```

```python
#how to delete columns by name
#we use drop() and axis=1
#delete a single column from the DataFrame
df1 = df.drop(labels=["group"], axis=1)
df1
```

```python
#delete a column by column number
df2 = df.drop(columns=df.columns[2])
df2
```

```python
#delete rows by index value
df3 = df.drop(labels=[1,15,20], axis=0)
df3
```

```python
#delete rows by index value in some range
df4 = df.drop(labels=range(40, 45), axis=0)
df4
```

```python
#delete a row by row number
df5 = df.drop(df.index[1])
df5
```

```python
#check the initial shape of the DataFrame
df.shape
```

```python
#delete rows where case numbers are zero
#this deletion is completed by "selecting" rows where total marks aren't equal 9
df6 = df.loc[df["mark_total"] != 9]
df6.shape
```

```python
#delete rows using slicing method
df.shape
df7 = df[:10]   #delete everything but the first 10 rows.
df7.shape
```

```python
#adding a column using insert()
df7.insert(1, "Age", [21, 20, 19, 18, 20, 21, 19, 20, 19, 18], True)
df7
```

```python
#copying data using new variables
new = df
new["new_column"]='changed'
new
```

```python
#copying data using copy() method
new = df.copy()
new["new_column"] = 'changed'
new
#difference with using new variables is that copy() method doesn't change df, it only changes new
```
